```python
import pandas as pd
import re
import numpy as np
```

```python
from opencc import OpenCC
cc1 = OpenCC('t2s') # 번체(표준) => 간체
cc2 = OpenCC('tw2s') # 번체(대만 표준) => 간체
cc3 = OpenCC('hk2s') # 번체(홍콩 변형) => 간체
```

```python
data = pd.read_excel('./데이터셋/foreign.xlsx')
```

```python
# CN 데이터의 제품명만 추출
data_cn = data[data['국가'] == 'CN'][['제품명']]
data_cn
```

```python
# 번역을 토대로 새로운 제품명 컬럼 생성
data_cn['번체 표준'] = data_cn['제품명'].apply(lambda x: cc1.convert(str(x))) #'t2s'
data_cn['번체 대만 표준'] = data_cn['제품명'].apply(lambda x: cc2.convert(str(x))) #'tw2s'
data_cn['번체 홍콩 표준'] = data_cn['제품명'].apply(lambda x: cc3.convert(str(x))) #'hk2s'
```
