```python
# 한글 맞춤법 검사
def split_text_by_language(text):
    korean = ' '.join(re.findall(r'[가-힣]+', text))
    english = ' '.join(re.findall(r'[a-zA-Z]+', text))
    return korean.strip(), english.strip()
```

```python
# 영어 맞춤법 검사
import language_tool_python

tool = language_tool_python.LanguageTool('en-US')

def check_english(text):
    matches = tool.correct(text)
    return matches
```

```python
# 제품명과 용량 분류
# 1차 전처리: 제품명에 숫자,괄호 분리
main['용량'] = main['제품명'].str.extract(r'(\([^)]+\)|\d+\s*[a-zA-Z]+)')
main['제품명'] = main['제품명'].str.replace(r'(\([^)]+\)|\d+\s*[a-zA-Z]+)', '', regex=True).str.strip()
```

```python
# 2차 전처리: 분류된 약품에 따라 제형을 정의
#1. 제품명의 마지막 글자에 '정'이 들어가면 정제형으로 변환하기
#2. 제품명의 마지막 글자에 '캡슐'이 들어가면 경질캡슐로 변환하기
# 약품 형태는 [액제, 정제형, 플라스타, 경질캡슐, 크림, 시럽제, 점안액, 연고, etc]

def infer_dosage(name):
  name = str(name).strip()
  if name.endswith('정'): 
    return '정제형'
  elif name[-2:] == '캡슐':
    return '경질캡슐'
  elif name[-2:] == '시럽':
    return '시럽제'
  elif name[-2:] == '연고':
    return '연고'
  elif name.endswith('겔'):
    return '겔제'
  elif name[-2:] == '크림':
    return '크림'
  elif name[-4:] == '플라스타': 
    return '플라스타'
  elif name[-3:] == '점안액':
    return '점안액'
  elif name[-1:] == '액':
    return '액제'

# etc는 수기 처리

main['제형'] = main['제품명'].apply(infer_dosage)
```
