```python
list1 = ['아모잘탄','트윈스타', '노바스크','올메텍','엑스브이','자누메트','트라젠타','제미메트',
         '자디앙','포시가','로수젯','리피토','크레스토','아토젯','리바로','다비가트란','엘리퀴스',
         '베르무트','토르세미드',"카르베딜롤", '니트로글리세린','프롤로정','딜라트렌','코자','아스피린',
         '플라빅스','타실렉스','리스크론','알로푸리놀','페북소스타',"콜키신",'세레타이드디스커스',
         '심비코트','몬테락','졸레어주','세레콕시브','몰루세트','아셈','레마토렉스','포사맥스플리스디',
         '에디로넬','테리본','악토넬']
```

```python
df1 = df[
    df['제품명A'].astype(str).apply(lambda x: any(p in x for p in list1)) |
    df['제품명B'].astype(str).apply(lambda x: any(p in x for p in list1))
]
df1
```
