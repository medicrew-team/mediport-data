```python
import pandas as pd

# 파일 불러오기
file_path = "./데이터셋/통합의약품.xlsx"
xls = pd.ExcelFile(file_path)

# 모든 시트 읽기
sheets = {}
for sheet_name in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    # 구매처 = '전문' 제거
    if '구매처' in df.columns:
        df = df[df['구매처'] != '전문']
    sheets[sheet_name] = df

# JSON 변환 (시트별로 dict)
json_data = {sheet: df.to_dict(orient="records") for sheet, df in sheets.items()}

output_path = "/mnt/data/통합의약품.json"
import json
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=2)

output_path
```
